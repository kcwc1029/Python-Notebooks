字典是一種無序的鍵（key）與值（value）的集合。想像一下，它就像一本字典一樣，你可以透過唯一的「詞彙」（鍵）來找到它所對應的「解釋」（值）。

![upgit_20260418_1776509202.png|498x278](https://raw.githubusercontent.com/kcwc1029/obsidian-upgit-image/main/2026/04/upgit_20260418_1776509202.png)


## 建立字典

````python
didi = {
    "A":"apple"
}
print(didi)
````

````python
# 建立一個名為 dict_empty 的空字典
dict_empty = {}
print(dict_empty)
````

````python
# 建立一個名為 fruits 的字典，鍵是水果名稱，值是水果的價格
fruits = {'西瓜': 15, '香蕉': 20, '水蜜桃': 25, "芒果":30}
print(fruits)
````

## 存取字典中的值

````python
fruits = {'西瓜': 15, '香蕉': 20, '水蜜桃': 25}

# 透過鍵 '西瓜' 來存取它的值
print(fruits['西瓜'])

# 透過鍵 '香蕉' 來存取它的值
print(fruits['香蕉'])

# ...
print(fruits['水蜜桃'])
````

````python
D = {
    "A":"apple",
    "B":"banana"
}

print(D)
D["C"] = "cake"
print(D)
````

## 新增鍵值對

````python
fruits = {'西瓜': 15, '香蕉': 20, '水蜜桃': 25}
# 新增一個新的鍵值對 '蘋果': 18
fruits['蘋果'] = 18
print(fruits)
````

## 刪除字典元素

可以使用 del 關鍵字來刪除特定的鍵值對。

![upgit_20260418_1776513484.png|498x278](https://raw.githubusercontent.com/kcwc1029/obsidian-upgit-image/main/2026/04/upgit_20260418_1776513484.png)


````python
fruits = {'西瓜': 15, '香蕉': 20, '水蜜桃': 25}

del fruits['香蕉'] # 刪除鍵 '香蕉' 以及它所對應的值
print(fruits)
````

````python
fruits = {'西瓜': 15, '香蕉': 20, '水蜜桃': 25, '芒果': 25}
fruits.pop('香蕉')
print(fruits)
````

## 遍歷字典

![upgit_20260418_1776513629.png|498x278](https://raw.githubusercontent.com/kcwc1029/obsidian-upgit-image/main/2026/04/upgit_20260418_1776513629.png)


````python
### 寫一個陣列
arr = ["A", "B", "C", "D", "E"]
for i in arr:
  print(i)
````

````python
players = {'西瓜': 15, '香蕉': 20, '水蜜桃': 25, '芒果': 25}

# key
for i in players.keys():
  print(i)


````

````python
### items()：同時遍歷鍵和值
players = {'西瓜': 15, '香蕉': 20, '水蜜桃': 25, '芒果': 25}


# 使用 items() 遍歷，同時取得鍵和值
print("--- 遍歷鍵和值 ---")
for name, team in players.items():
    print(f"{name} 的球隊是 {team}")
````

````python
### keys()：遍歷字典的鍵
fruits = {'西瓜': 15, '香蕉': 20, '水蜜桃': 25, '芒果': 25}
for i, j in fruits.items():
  print(i, j)
````

````python
### values()：遍歷字典的值
players = {
    'Stephen Curry': 'Golden State Warriors',
    'Kevin Durant': 'Golden State Warriors'
}

# 使用 values() 遍歷，只取得值
print("--- 遍歷值 ---")
for team in players.values():
    print(f"球隊名稱：{team}")
````

## sorted()：排序

````python
fruits = {'Orange': 60, 'Apple': 100, 'Grape': 80}

# 對key排序（依照字母順序）
sorted_items_by_key = sorted(fruits.items())
print(sorted_items_by_key)
# 對value排序
sorted_items_by_value = sorted(fruits.items(), key=lambda item: item[1])
print(sorted_items_by_value)
# 這邊用到的lambda，我們會在【函數】再來討論~~~
````

## 字典其餘的常用方法



````python
### clear()：清空所有元素
fruits = {'西瓜': 15, '香蕉': 20, '水蜜桃': 25}
fruits.clear()
print(fruits)
````

````python
### in檢查鍵是否存在
fruits = {'西瓜': 15, '香蕉': 20, '水蜜桃': 25}

if '香蕉' in fruits: print('字典裡有香蕉')
else: print('字典裡沒有香蕉')

if '鳳梨' in fruits: print('字典裡有鳳梨')
else: print('字典裡沒有鳳梨')
````

````python
### update()合併字典
# update() 方法可以將一個字典的鍵值對添加到另一個字典中。如果兩個字典有相同的鍵，則被合併的字典（參數）中的值會覆蓋掉原字典中的值。
dealerA = {'1': 'Nissan', '2': 'Toyota', '3': 'Lexus'}
dealerB = {'1': 'BMW', '2': 'Benz'}

# 將 dealerB 的內容合併到 dealerA
dealerA.update(dealerB)
print(dealerA)
````
