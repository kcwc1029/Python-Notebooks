## type()：查詢資料型態
在開始學不同資料型態前，我們需要一個工具來確認我們手上拿的到底是什麼。Python 提供了一個超好用的`type()` 函式。

```python
score = 95 # 整數 (int)
print(type(score))
# <class 'int'>

pi_value = 3.14159 # 浮點數 (float)
print(type(pi_value))
# <class 'float'>

slogan = "顏面神經失調" # 字串 (str)
print(type(slogan))
# <class 'str'>
```

## Integer

Def. 就是沒有小數點的數字，正數、負數、零都可以。
在 Python 3 中，int 沒有儲存大小限制！你可以存一個天文數字，只要你的電腦記憶體夠大就行。

> 在其他程式語言中，整數通常會有儲存空間大小限制（例如 32 位元整數範圍：-2,147,483,648 到 2,147,483,647）。

## float
帶有小數點的數字，例如 1.0、48.52、−0.001。
> 在其他程式語言中，可以再進一步分為【單經度浮點數float】與【雙經度浮點數double】。
> 在 Python 中，其實只有一種浮點數類型：float，它對應的是雙精度浮點數（double precision）。

## 類型自動轉換

```python
# 情境：我們有 5 顆蘋果，每顆重量是 2.5 公斤。
apple_count = 5       # int (整數)
price_per_kg = 2.5    # float (浮點數)

total_weight = apple_count * price_per_kg # 運算結果是 12.5 (自動從 int 升級為 float)

print(total_weight)
print(type(total_weight))
```

## 強制資料型態轉換
在程式設計中，資料型態的轉換是家常便飯。
例如：使用者輸入的數字其實是字串，你需要把它轉成整數才能進行數學運算。

![upgit_20260420_1776665883.png|329x94](https://raw.githubusercontent.com/kcwc1029/obsidian-upgit-image/main/2026/04/upgit_20260420_1776665883.png)


```python
# 1. 數值之間的轉換
x = 10.5
print(f"10.5 轉為整數: {int(x)}")

# 2. 字串轉整數/浮點數 (前提是字串內容看起來像數字)
age_str = "18"
weight_str = "65.5"

# 轉型後才能做數學運算
next_age = int(age_str) + 1
print(f"明年年紀是: {next_age}") # 輸出: 19
print(f"體重增加 2 公斤: {float(weight_str) + 2}") # 輸出: 67.5

# 3. 數值轉字串 (在串接文字時非常重要)
pi = 3.14
message = "圓周率是 " + str(pi) # 數字不能直接跟字串相加，必須先轉成字串
print(message)
```

### 寄件與收件的兩大動作：打包與拆包
![upgit_20260420_1776666205.png|329x184](https://raw.githubusercontent.com/kcwc1029/obsidian-upgit-image/main/2026/04/upgit_20260420_1776666205.png)


```python
name = '我今年3歲'
print(f"原始字串長度 (字元數): {len(name)}") # 輸出: 5

# 步驟 1: 字串 'str' -> 二進位 'bytes' (打包/編碼)
name_bytes = name.encode('utf-8')
print(f"打包後: {name_bytes}")
print(f"bytes 長度 (位元組數): {len(name_bytes)}") 

# 步驟 2: 二進位 'bytes' -> 字串 'str' (拆包/解碼)
name_unicode = name_bytes.decode('utf-8')
print(f"拆包後的字串: {name_unicode}")
```
