## Problem. 確認資料型態
題目描述：請建立三個變數，分別儲存學生的分數、身高與姓名，並印出它們的資料型態。
變數名稱與輸入設定：

* score：數值設定為 88
* height：數值設定為 175.5
* student_name：數值設定為 "小明"
* 輸出要求：依序使用 type() 函式印出這三個變數的資料型態。

## Problem. 數值運算與型態自動轉換
請計算買水果的總花費，並觀察 Python 如何自動處理資料型態的轉換。

變數名稱與輸入設定：
* apple_count：數值設定為 8
* price_per_apple：數值設定為 15.5
* total_price：設定為 apple_count 乘以 price_per_apple

輸出要求：
* 第一行印出 total_price
* 第二行印出 total_price 的資料型態

## Problem. 字串轉數字的強制轉換
模擬接收到字串格式的資料，請將其轉換成正確的數值型態後進行數學計算。

變數名稱與輸入設定：
* id_str：數值設定為 "1234" (字串型態)
* base_number：數值設定為 500
* final_password：將 id_str 轉為整數型態後，加上 base_number

輸出要求：印出 final_password 的數值。

## Problem. 其他進位制字串轉十進位
請將指定的二進位字串與十六進位字串，透過指定進位制參數轉換為十進位整數後相加。

變數名稱與輸入設定：
* bin_str：數值設定為 "1010"
* hex_str：數值設定為 "F"
* total_decimal：將 bin_str (以二進位讀取) 與 hex_str (以十六進位讀取) 轉為十進位數值後相加

輸出要求：印出 total_decimal 的數值。

## Problem. 字串的乘法與加法

## Problem. 資料的打包與拆包
請將中文字串打包成二進位資料 (bytes)，觀察其長度變化，再將其拆包轉回正常字串。

變數名稱與輸入設定：
* secret_msg：數值設定為 "機密任務"
* encoded_msg：將 secret_msg 以 utf-8 編碼格式進行打包 (encode)
* decoded_msg：將 encoded_msg 解碼 (decode) 轉回字串

輸出要求：
* 第一行使用 len() 函式印出 encoded_msg 的位元組長度
* 第二行印出 decoded_msg 的內容
