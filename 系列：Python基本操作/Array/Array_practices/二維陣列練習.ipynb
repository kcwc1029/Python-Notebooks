### Problem. 全班總分掃描
請建立一個包含四位學生資料的二維陣列 `class_data`：
-   `["A", 60, 70, 80, 0]`
-   `["B", 90, 90, 90, 0]`
-   `["C", 50, 60, 50, 0]`    
-   `["D", 80, 75, 70, 0]`

請使用 `for` 迴圈遍歷（走訪）整個 `class_data`。在迴圈中，利用 `sum()` 搭配切片，自動計算每位學生的總分（索引 1 到 3 的總和），並填入索引 `4` 的位置。最後印出完整的陣列。

輸出內容：
```
全班總分計算完成：
[['A', 60, 70, 80, 210], ['B', 90, 90, 90, 270], ['C', 50, 60, 50, 160], ['D', 80, 75, 70, 225]]
```

```python=

    ["A", 60, 70, 80, 0],
    ["B", 90, 90, 90, 0],

for student in class_data:
```
