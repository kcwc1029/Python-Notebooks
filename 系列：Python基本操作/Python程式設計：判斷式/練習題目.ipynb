## Problem. 單一條件判斷
請設計一個程式，判斷使用者的錢包餘額是否足夠購買一杯 65 元的珍珠奶茶。

變數名稱與輸入設定：
* wallet_money：使用 input() 讀取輸入金額，並轉換為整數
* boba_price：數值設定為 65

輸出要求：如果 wallet_money 大於或等於 boba_price，印出「餘額充足，可以購買」；否則印出「餘額不足」。

<!--
wallet_money = int(input("請輸入錢包餘額："))
boba_price = 65

if wallet_money >= boba_price:
    print("餘額充足，可以購買")
else:
    print("餘額不足")
 -->

```python
wallet_money = int(input())
boba_price = 65
if wallet_money > boba_price:
  print("餘額充足，可以購買")
else:
  print("餘額不足")
```

## Problem. 奇數與偶數判斷

請判斷使用者輸入的數字是奇數還是偶數 (利用除以 2 的餘數來判斷)。

變數名稱與輸入設定：
* number：使用 input() 讀取數字並轉為整數

輸出要求：若 number 除以 2 的餘數等於 0，印出「這是偶數」；否則印出「這是奇數」。

<!--
number = int(input("請輸入一個整數："))

if number % 2 == 0:
    print("這是偶數")
else:
    print("這是奇數")
 -->

## Problem. 多重條件判斷

* age：使用 input() 讀取年紀並轉為整數

age = int(input("請輸入年紀："))

    print("請購買全票")

    print("請購買敬老票")
 -->

## Problem. None 型態判斷
模擬系統抓取使用者的個人檔案照片，有時候使用者可能尚未上傳照片，此時變數會是 None。請撰寫程式檢查這個狀態。

變數名稱與輸入設定：
* profile_picture：數值設定為 None

輸出要求：使用 if 判斷 profile_picture 是否等於 None (或使用 is None)，若是則印出「使用者尚未上傳照片」，否則印出「照片讀取成功」。

<!--
profile_picture = None

if profile_picture == None:
    print("使用者尚未上傳照片")
else:
    print("照片讀取成功")
 -->
