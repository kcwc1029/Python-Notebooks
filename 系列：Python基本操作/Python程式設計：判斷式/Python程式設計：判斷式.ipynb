## 邏輯運算子 (Logical Operators)

```python
print(10<8) # False
print(not(10<8)) # True
print(True and False) # False
print(True or False) # True
print(not(True and False)) # True
print(not(True or False)) # False
```

```python
### 判斷分數
score = input("請輸入分數:")
sc = int(score)
if (sc >= 90):
    print("A")
elif (sc >= 80):
    print("B")
elif (sc >= 70):
    print("C")
elif (sc >= 60):
    print("D")
else:
    print("F")
```

```python
### 電影票價計算機 (結合 or 與字串判斷)
age = int(input("請輸入您的年齡: "))
is_student = input("您是學生嗎？(請輸入 Y 或 N): ")

# 只要小於 6 歲「或」大於等於 65 歲，就能買優待票
if age < 6 or age >= 65:
    print("您可以購買優待票：150元")
# 如果不是優待身分，但字串判斷為學生 (轉換為大寫避免大小寫輸入差異)
elif is_student.upper() == "Y":
    print("您可以購買學生票：200元")
# 其他情況都是全票
else:
    print("全票：250元")
```

```python
### 閏年判斷 (結合 and、or 與數學運算)
year = int(input("請輸入西元年份: "))

# 條件一：是 4 的倍數「且」不是 100 的倍數
# 條件二：是 400 的倍數
# 只要符合條件一「或」條件二，就是閏年
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} 年是閏年！")
else:
    print(f"{year} 年是平年。")
```

```python
### BMI 健康狀態評估 (連續範圍判斷)
height_cm = float(input("請輸入身高 (公分): "))
weight_kg = float(input("請輸入體重 (公斤): "))

# 將身高轉為公尺並計算 BMI
height_m = height_cm / 100
bmi = weight_kg / (height_m ** 2)
print(f"您的 BMI 值為: {bmi:.2f}")

if bmi < 18.5:
    print("狀態：體重過輕")
# Python 支援像數學一樣的連續寫法：18.5 <= bmi < 24
elif 18.5 <= bmi < 24:
    print("狀態：正常範圍")
elif 24 <= bmi < 27:
    print("狀態：過重")
else:
    print("狀態：肥胖")
```

```python
### 百貨公司折扣計算 (注意條件順序)
# 在撰寫 if...elif...else 時，條件的順序非常重要，程式會從上往下檢查，只要符合其中一個，後面的就不會再看了。
total = int(input("請輸入消費總金額: "))

# 必須把最嚴格的條件 (金額最大的) 放在最前面
if total >= 10000:
    print("符合滿萬打 8 折優惠！")
    final_price = total * 0.8
elif total >= 5000:
    print("符合滿五千打 9 折優惠！")
    final_price = total * 0.9
else:
    print("未達折扣門檻。")
    final_price = total

print(f"折扣後應付金額為: {int(final_price)} 元")
```
