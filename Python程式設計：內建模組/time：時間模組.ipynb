在程式設計中，我們經常需要處理時間，例如：記錄事件發生的時間、計算一段程式碼執行了多久、或是以特定格式顯示日期。time 模組就是為此而生。

## time() 和 process_time(): 計算程式執行時間
* time.time(): 會回傳從「Epoch」（一個標準的參考時間點：1970年1月1日 00:00:00）到現在所經過的總秒數（一個浮點數）。我們可以透過記錄開始和結束的秒數，來計算程式執行的牆上時鐘時間 (Wall Clock Time)，也就是我們實際感受到的流逝時間。
* time.process_time(): 只會計算 CPU 執行目前程式所花費的時間，不包含程式等待（例如等待使用者輸入）的時間。這在評估演算法本身的效率時特別有用。

```python
# 計算圓周率所花費的 CPU 時間
import time

# 記錄 CPU 開始時間
start_time = time.process_time()

# 進行一個需要大量計算的任務
pi = 0
for i in range(10000000): # 迴圈執行一千萬次
    pi += 4 * ((-1)**i) / (2*i + 1)

# 記錄 CPU 結束時間
end_time = time.process_time()

# 印出結果和花費的 CPU 時間
print("Pi = ", pi)
print(f"總共花費了 {end_time - start_time:7.4f} 秒") # 格式化輸出
```

## localtime() 和 asctime(): 獲取當前時間
time.localtime(): 會回傳一個包含當前本地時間資訊的特殊元組 (tuple)，稱為 struct_time。它把年、月、日、時、分、秒等資訊分開存放，方便我們單獨取用。

time.asctime(): 直接回傳一個格式化好的、人類易於閱讀的星期、月份、日期、時間和年份的字串。

* 索引 0 (tm_year): 年
* 索引 1 (tm_mon): 月 (1-12)
* 索引 2 (tm_mday): 日 (1-31)
* 索引 3 (tm_hour): 時 (0-23)
* 索引 4 (tm_min): 分 (0-59)
* 索引 5 (tm_sec): 秒 (0-59)
* 索引 6 (tm_wday): 星期幾 (0=週一, 6=週日)
* 索引 7 (tm_yday): 一年中的第幾天 (1-366)
* 索引 8 (tm_isdst): 是否為日光節約時間

```python
import time

# 獲取本地時間的 struct_time 物件
xtime = time.localtime()

# 分別印出各個欄位
print("xtime = ", xtime)
print("年 = ", xtime.tm_year) # 也可以用索引 xtime[0]
print("月 = ", xtime.tm_mon)  # 也可以用索引 xtime[1]
print("日 = ", xtime.tm_mday) # 也可以用索引 xtime[2]
print("時 = ", xtime.tm_hour) # 也可以用索引 xtime[3]
```

## strftime(): 格式化時間輸
strftime() (string format time) 是最強大、最靈活的時間格式化函式。您可以自訂時間的輸出格式。

如果省略「時間元組」，則預設使用當前的 localtime()
常用格式代碼：
* %Y: 四位數的年份 (例如 2025)
* %m: 兩位數的月份 (01-12)
* %d: 兩位數的日期 (01-31)
* %H: 24小時制的兩位數小時 (00-23)
* %M: 兩位數的分鐘 (00-59)
* %S: 兩位數的秒數 (00-59)

```python
import time

# 定義我們想要的輸出格式
formatted_time = time.strftime("%Y-%m-%d %H:%M:%S", time.localtime())
print(formatted_time)

now = time.strftime("【%Y/%m/%d】 %H:%M:%S")
print(f"目前顯示時間：{now}")
```

## 實作：牆上時鐘 vs. CPU 耗時 (time vs. process_time)
撰寫一個程式，在執行過程中加入 `time.sleep(2)`（讓程式暫停 2 秒）。分別紀錄並印出這段時間內「實際流逝的時間」與「CPU 執行的時間」。

```python
import time

# 紀錄開始
wall_start = time.time()
cpu_start = time.process_time()

print("程式進入睡眠 2 秒...")
time.sleep(2) # 暫停執行

# 紀錄結束
wall_end = time.time()
cpu_end = time.process_time()

print(f"實際流逝時間 (Wall Clock): {wall_end - wall_start:.4f} 秒")
print(f"CPU 執行時間 (Process Time): {cpu_end - cpu_start:.4f} 秒")
```
