隨著程式越寫越多，起初把所有的功能、類別都寫在同一個 .py 檔案裡，但很快這個檔案就變得數千行長，難以閱讀、修改和維護，而模組 (Module) 就是為了解決這個問題而誕生的。

在 Python 中，任何一個 .py 檔案都可以被視為一個模組。將相關的函式、類別和變數組織在一個獨立的檔案中，以達到程式碼的組織化與重複使用。


![upgit_20260420_1776671930.png|329x184](https://raw.githubusercontent.com/kcwc1029/obsidian-upgit-image/main/2026/04/upgit_20260420_1776671930.png)

> 一個模組就像一個「工具箱」。例如一個專門放「美食相關」工具的 makefood.py 工具箱，或是建立一個專門放「銀行相關」工具的 banks.py 工具箱。
當您需要用到某個工具時，只需要「匯入 (import)」對應的工具箱即可，而不需要把所有工具都散落在同一個地方。

## 建立模組檔案
我們來建立一個名為 makefood.py 的檔案。這個檔案本身就是我們的模組。

````python
%%writefile makefood.py
def make_icecream(toppings):
    """ 列出製作冰淇淋的配料 """
    print("製作冰淇淋的配料如下：")
    for topping in toppings:
        print("--- ", topping)

# 定義製作飲料的函式
def make_drink(size, drink):
    """ 輸入飲料規格與種類，然後輸出飲料 """
    print("--- 客製化飲料 ---")
    print("尺寸：", size)
    print("飲料：", drink)
````

## 在其他程式中應用自己建立的函式
```
├── makefood.py
└── main.py
```

````python
### 方法一：導入整個模組 import 模組名稱
# 這種方式的好處是，程式碼非常清楚，任何人一看就知道 make_icecream 這個函式是來自 makefood 模組，不會搞混。
# main.py
import makefood

# 呼叫函式時，必須在前面加上 "模組名稱."
makefood.make_icecream(['草莓醬', 'OREO餅乾', '巧克力碎片'])
makefood.make_drink('large', 'coke')
````

````python
### 方法二：從模組中導入特定函式 from ... import ...
# main.py
from makefood import make_icecream, make_drink

# 因為已經明確導入，可以直接呼叫函式名稱
make_icecream(['草莓醬', 'OREO餅乾', '巧克力碎片'])
make_drink('large', 'coke')
````

````python
### 方法三：使用 as 給模組或函式取一個「別名」
# 有時候模組名稱太長，或是導入的函式名稱與現有程式碼衝突，我們可以用 as 來取一個好記又簡短的別名。
# *	語法 (模組別名)：import 模組名稱 as 別名
# *	語法 (函式別名)：from 模組名稱 import 函式名稱 as 別名

# main.py
import makefood as m  # 將 makefood 取一個簡短的別名 m

m.make_icecream(['草莓醬'])
m.make_drink('small', 'tea')
````

## if __name__ == '__main__'
我們可以把 if __name__ == '__main__': 想像成一道「身分驗證門」，這道門的作用是：判斷這個 Python 檔案現在是「主角」還是「配角」。

![upgit_20260420_1776672631.png|329x172](https://raw.githubusercontent.com/kcwc1029/obsidian-upgit-image/main/2026/04/upgit_20260420_1776672631.png)

### 為什麼需要這道門？
假設你寫了一個好用的工具模組 my_tools.py，並在最下方留了一行測試代碼：

````python
%%writefile my_tools.py
### 先建立一個模組
def add(a, b):
    return a + b

# 順手寫來測試用的代碼
print("測試一下：1+1=", add(1, 1))
# 直接執行 my_tools.py 時，印出測試結果很正常
````

````python
# 但如果隔天你在新程式 main_app.py 中借用了它：
# 檔案名稱：main_app.py
import my_tools

print("歡迎來到我的新程式！")
# 這時螢幕會先印出「測試一下：1+1= 2」，然後才印出歡迎詞。這是因為 Python 在執行 import 時，會自動把借來的檔案從頭到尾跑一次，導致那句測試代碼跑出來搗亂。
````

為了解決這個尷尬的情況，Python 給每個檔案配發了一張隱藏身分證，叫做 __name__：

* 當你「直接執行」檔案時：身分證會自動設定為 '__main__'（代表：我是今天的主角！）。
* 當你「借用 import」檔案時：身分證會設定為它的檔案名稱，例如 'my_tools'（代表：我只是來幫忙的配角）。

````python
%%writefile my_tools.py
def add(a, b):
    return a + b

# 設立身分驗證門
if __name__ == '__main__':
    print("測試一下：1+1=", add(1, 1)) # 只有當我是主角（被直接執行）時，門才會打開
````

````python
# 但如果隔天你在新程式 main_app.py 中借用了它：
# 檔案名稱：main_app.py
import my_tools

print("歡迎來到我的新程式！")
````
