## 輸出
### 格式化輸出format()
使用 `{}` 作為佔位符，並在 format() 函數的參數中提供要填入的內容。
```py
# 使用方式
print("格式化字串 {}".format(變數))
```

````python
# 使用範例
s = "Hello"
x = 99
print("字串 = {}".format(s))
print("數字 = {}".format(x))
print("字串 = {}，數字 = {}".format(s, x))
````

### 格式化字串 f-string
在字串前面加上 f，並在 `{}` 佔位符中直接放入變數或運算式，是目前最推薦的格式化方式。
```py
# 使用方式
變數 = "值"
print(f"格式化字串 {變數}")
```

````python
# 使用範例
name = "小明"
message = "Hello"
print(f"姓名：{name}，訊息：{message}")
````

## input()：輸入 
input() 函數用於從鍵盤讀取使用者輸入的資料。
```py
# 基本用法
value = input("prompt:")
# prompt 是提示訊息，會顯示在螢幕上，告訴使用者需要輸入什麼。
# value 是一個變數，用於儲存使用者輸入的內容。
# input() 函數讀取到的所有資料，無論看起來像數字還是文字，都會被當作字串（string）來處理。如果你需要對數字進行數學運算，必須使用 int() 或 float() 等函數將其轉換為整數或浮點數。
```

````python
print("歡迎使用成績輸入系統")
name = input("請輸入姓名：")
engh = input("請輸入英文成績：")
math = input("請輸入數學成績：")
total = int(engh) + int(math) # 將字串轉換為整數後相加
````

## 處理字串的數學運算：eval() 
eval() 函數可以將一個字串當作 Python 的運算式來執行，並返回運算結果。在處理使用者輸入的數學公式時非常有用。

````python
# numberStr = input("請輸入數值公式：")
numberStr = "5 * 9 + 10"
result = eval(numberStr)
print(f"計算結果：{result}")
````

## 輸出資料到檔案 open()
在 Python 中，所有檔案操作都從 open() 函數開始。

| 程式碼 | 說明 |
| --- | --- |
| `"out.txt"` | 檔案名稱 |
| `mode="w"` | 存取模式 (Mode): `w` 代表 寫入 (write) 模式。如果 `out.txt` 檔案不存在，會創建一個新檔案；如果檔案已存在，會覆蓋 (清除) 掉舊的內容。 |
| `encoding="utf-8"` | 編碼方式: 指定檔案內文字的編碼。`utf-8` 是目前最通用的編碼，能良好支援中文、英文等多國語言。 |
| `as f:` | 將開啟後得到的檔案物件 (File Object) 賦值給變數 `f` 。 |

這種方式的問題是：如果程式在 f.write() 步驟發生錯誤，f.close() 就可能不會被執行，導致檔案資源沒有釋放，甚至資料遺失或損壞。

這邊會推薦`with open(...) as f:` 語法，這是 Python 處理檔案的標準且最安全的方法。


````python
with open("out.txt", mode="w", encoding="utf-8") as f:
    f.write("好熱歐今天")
    # print(text01, file=f)
````
